## Plotting Fault Slip and Length   

You've measured a the lengths and amounts of slip on various faults in Figure 7 of Underhill and Stoneley, 1998.  Let's go ahead and plot them.

```python
# first we need to import some useful functions / software
# to do so, we run this cell---we click in the cell and press the run button above

import matplotlib.pyplot as plt
from matplotlib.pyplot import plot
from numpy import array

print('You have imported some useful software')
```

```python
# list all the pairs of fault length and slips
# prs = [(length1, slip1),(length2, slip2), (length3, slip3), ...]
prs = [(1,1),(2,2),(3,3)]

print('The pairs of fault length and slip are ')
print(prs)
```

```python
# extract the lengths and slips to separate lists

# all the fault lengths
lengths = [pr[0] for pr in prs]

#  all the fault slips
slips = [pr[1] for pr in prs]

# and plot
f = plt.figure(figsize=(12,10))

plot(lengths,slips,marker='o',linestyle='none',markersize=8);
plt.xlabel('length (km)',fontsize='x-large');
plt.ylabel('slip in two-way travel time (s)',fontsize='x-large');
```
